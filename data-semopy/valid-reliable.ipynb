```python
num = raw.loc[:,'X111':'Y224']
col = num.columns
num = np.select([
    num=='Sangat Setuju',
    num=='Setuju',
    num=='Netral',
    num=='Tidak Setuju',
    num=='Sangat Tidak Setuju'
    ],[5,4,3,2,1],np.nan)
df = pd.DataFrame(num,columns=col)
df.fillna(inplace=True,method='ffill')

np.unique(df,return_counts=True)
df.shape
```

```python
vld = pd.DataFrame(index=df.loc[:,"X111":"Y224"].columns, columns=['R','against'])
df['AvX11'] = df.loc[:,"X111":"X115"].mean(axis=1).round(2)
df['AvX12'] = df.loc[:,"X121":"X128"].mean(axis=1).round(2)
df['AvX13'] = df.loc[:,"X131":"X136"].mean(axis=1).round(2)

df['AvX21'] = df.loc[:,"X211":"X213"].mean(axis=1).round(2)
df['AvX22'] = df.loc[:,"X221":"X223"].mean(axis=1).round(2)
df['AvX23'] = df.loc[:,"X231":"X233"].mean(axis=1).round(2)
df['AvX24'] = df.loc[:,"X241":"X243"].mean(axis=1).round(2)

df['AvY11'] = df.loc[:,"Y111":"Y113"].mean(axis=1).round(2)
df['AvY12'] = df.loc[:,"Y121":"Y123"].mean(axis=1).round(2)
df['AvY21'] = df.loc[:,"Y211":"Y219"].mean(axis=1).round(2)
df['AvY22'] = df.loc[:,"Y221":"Y224"].mean(axis=1).round(2)


for i in vld.index:
    vld.loc[i,"R"] = round( stats.pearsonr(df[i], df[f'Av{i[:3]}'])[0],3) 
    vld.loc[i,"against"] = f'Av{i[:3]}'
vld.to_excel('validitas_kor_koef_per_item.xlsx')
# vld.head()
```

```python
# x = np.unique(X11,return_counts=True)
# x[1][np.where(x[0] == 2)[0][0]]

def getIt(x,i):
 y= [np.where(x[0] == i)]
 z = x[1][y][0][0]
 return 0 if len(z) == 0 else z[0]

def getPercentage(x,i):
 y = [np.where(x[0] == i)]
 z = x[1][y][0][0]
 c = 0 if len(z) == 0 else z[0]
 return round(c / np.sum(x[1]) *100,2)

for d in range(1,6):
 reliability[d] = [getIt(np.unique(i,return_counts=True),d) for  i  in alldf2]
 reliability[f'{d} -%'] =  [getPercentage(np.unique(i,return_counts=True),d) for  i  in alldf2]  
# [[getIt(np.unique(i,return_counts=True),d) for  i  in alldf] for d in range(1,6)]
# [getIt(np.unique(i,return_counts=True)) for  i  in alldf]
# [1,2,3]

reliability['avg']  = [round(np.mean(np.mean(d)),2) for d in alldf2]

# reliability[:4]
reliability
```
